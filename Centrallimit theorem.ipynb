```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
```

```python
sns.displot(df["displacement"],kind = "kde")
```

```python
sample_mean_list = []
for i in(0,35):
    random_sample = df["displacement"].sample(30)
    sample_mean_list.append(random_sample.mean())
```

```python
sns.displot(sample_mean_list,kind = "kde")
```

```python
import statistics
statistics.mean(sample_mean_list)
```
